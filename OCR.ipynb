```python
import matplotlib.pyplot as plt
import keras_ocr
```

```python
pipeline=keras_ocr.pipeline.Pipeline()
```

```python
images=[keras_ocr.tools.read(img) for img in ['https://lh6.googleusercontent.com/qEpK_Ls4sy7ELYmrNf28sc765f6mEFYABWlzphFWx-OTQXcmOiFLYEjHynqRak6aS0sJP4p1Ca8evL9ks01Zs9Rtbo07ZRI9v3JJ9LKE',
                                              'https://lh6.googleusercontent.com/bZ1GboVVbM9lEk50E-36LFFmZhzSUJPzNw0vC9fzv6lhPZ9cHWinvr_6S8alyRAFbnrYFHuLMrp8PbWuiJ7C0WBEoC_AuFGOdbXjw84dEQ6uOyQmE7OaXnxq20CCWp32W9Mm77uK'] ]
```

```python
import numpy as np
print(np.shape(images))
```

```python
prediction=pipeline.recognize(images)
```

```python
prediction_groups = pipeline.recognize(images)
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
